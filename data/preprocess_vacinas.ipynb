```python
import os
import pandas as pd

INDIR = "/home/renato/LargeFiles/datavis/"
OUTDIR = "./vacina/"
```

```python
def get_percentage_vaccine_type(df):
    vacinas = df["vacina_nome"].groupby(df["vacina_nome"]).count()
    vacinas_sum = vacinas.sum()
    vacinas = vacinas.apply(lambda x: (x/vacinas_sum)*100)
    return vacinas

def get_percentage_vaccine_dose(df):
    dose = df["vacina_descricao_dose"].groupby(df["vacina_descricao_dose"]).count()
    dose_sum = dose.sum()
    dose = dose.apply(lambda x: (x/dose_sum)*100)
    dose = dose.rename(lambda x: x.strip().replace("ª", "").replace("\xa0", " "))
    dose = dose.rename({"Dose": "Única"})
    dose = dose.groupby(level=0).sum()
    dose = dose.reindex(index = ['1 Dose', '2 Dose', 'Única'])
    return dose
```

```python
STATES = ['am', 'ma', 'mg', 'ce', 'rs', 'pe', 'df', 'pr', 'sp']
series_type = []
series_dose = []
for state in STATES:
    df = pd.read_csv(os.path.join(INDIR, f"reduced-{state}.csv"))
    
    vacinas = get_percentage_vaccine_type(df)
    dose = get_percentage_vaccine_dose(df)
    
    vacinas = vacinas.rename(state.upper())
    dose = dose.rename(state.upper())
    
    series_type.append(vacinas)
    series_dose.append(dose)
```

```python
vacinas = pd.concat(series_type, axis=1)
vacinas.to_csv(os.path.join(OUTDIR, "vacinas.csv"))
vacinas
```

```python
doses = pd.concat(series_dose, axis=1)
doses.to_csv(os.path.join(OUTDIR, "doses.csv"), index=False)
doses
```
